# swmmx: all `set` functions

This notebook is a categorized reference for every public setter path in `swmmx`, including read-only paths that intentionally raise `ReadOnlyParameterError`.

Setter pattern:

```python
m.set.<main_category>.<sub_category>(value, ids=None)
```

- Object-level setters accept one scalar value, a 1D NumPy array, or a pandas Series.
- Scalars broadcast across all selected IDs; vector inputs must match the number of selected IDs.
- Structured fields such as coordinates, geometry, points, and row groups require one structured payload per selected ID.
- Reference fields validate target IDs where practical.
- Derived and result parameters are read-only and raise `ReadOnlyParameterError` if set.
- Use a cloned model when experimenting so the original input remains untouched.


````python
from pathlib import Path
from swmmx import swmm

example_path = Path('examples/example.inp')
m = swmm(example_path).clone()  # safe working copy for experiments
````

## Common setter patterns

```python
m.set.conduit.roughness(0.013)                          # broadcast to all conduits
m.set.conduit.roughness(0.014, ids='C1')                 # one conduit
m.set.conduit.roughness([0.013, 0.014], ids=['C1','C2']) # one value per selected ID
m.set.node.coordinate((100.0, 200.0), ids='J1')         # structured value
m.set.conduit.flow(1.0)                                 # read-only -> error
```

# swmmx: all `get` functions

This notebook is a categorized reference for every public getter available in `swmmx`.

Getter pattern:

```python
m.get.<main_category>.<sub_category>(ids=None, format=None)
```

- For object collections, `ids` can be `None`, one string ID, or a list of string IDs.
- `format=None` defaults to NumPy-style output; `format='df'` requests pandas output.
- A single explicit ID usually returns a scalar or one structured value.
- Multiple IDs or `ids=None` usually return arrays/DataFrames.
- Result variables require a completed run when matching objects exist.
- If a valid model has no objects of a requested type, an all-object getter returns an empty result.


````python
from pathlib import Path
from swmmx import swmm

example_path = Path('examples/example.inp')
m = swmm(example_path)

# Run the model before using result getters such as m.get.link.flow(...)
# when matching objects exist and you want actual simulation results.
# m.run()
````

## Common getter patterns

```python
m.get.conduit.length()                         # all conduits, NumPy output
m.get.conduit.length(ids='C1')                  # one conduit, scalar output
m.get.conduit.length(ids=['C1', 'C2'])          # selected conduits
m.get.link.flow(ids=['C1', 'C2'], format='df')  # result time series after m.run()
m.get.weir.crest_height()                       # empty array if the model has no weirs
```

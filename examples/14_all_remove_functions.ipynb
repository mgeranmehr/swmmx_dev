# swmmx: all `remove` functions

This notebook is a categorized reference for every public remove endpoint in `swmmx`.

Remove pattern:

```python
m.remove.<category>.<element_type>(ids, force=False)
```

- `ids` can be one string ID or a list of string IDs.
- By default, referenced objects are protected from unsafe removal.
- `force=True` only performs conservative cascades that are implemented safely; otherwise a clear error is raised.
- Successful removals return a summary dictionary and invalidate stale results.


````python
from swmmx import swmm

m = swmm(new='SI')
# Use a scratch model or clone before experimenting with removals.
````

## Common remove patterns

```python
m.remove.link.conduit('C1')
m.remove.node.junction(['J1', 'J2'])
m.remove.node.junction('J1', force=True)
```
